## Asking and Answering Questions

### Q.1) Which team has won most of the game at home in 2014?

First we will make a new column called `Winner` in which we decide which team has won the game(Home Team/Away Team).

```python
home_team_goals = list(match_raw_data['Home Team Goals'])
away_team_goals = list(match_raw_data['Away Team Goals'])
res=[]
for i in range(len(home_team_goals)):
    if(home_team_goals[i] > away_team_goals[i]):
        res.append('Home Team Wins')
    elif(home_team_goals[i ]< away_team_goals[i]):
        res.append('Away Team Wins')
    else:
        res.append('Tie')
```

```python
match_data['Winner'] = res
```

```python
match_data.loc[match_data['Winner']=='Home Team Wins']
```

```python
match_data.iloc[1]['Home Team Name']
```

```python
match_data[(match_data['Year']==2014) & (match_data['Winner']=='Home Team Wins')]
```

```python
home_win = match_data[(match_data['Year']==2014) & (match_data['Winner']=='Home Team Wins')]
sns.countplot(data=home_win,y='Home Team Name');
plt.title('Most Win At Home in 2014');
```

### Q.2) Till now how many matches ended with `Penalties`?

The matches which are labeled as `Tie` but ended with `Penalties`:

```python
match_data[(match_data['Winner']=='Tie') & (match_data['Win conditions']!=" ")]
```

Correcting of the data where match is played by `Penalties` and labeled as `Tie`.

```python
match_data[(match_data['Winner']=='Tie') & (match_data['Win conditions']!=" ")]
```

```python
x =re.findall('\d',penalties.iloc[0])
z=[int(i) for i in x]
type(z[1])
```

```python
def convert_to_int(x):
    temp = re.findall('\d',x)
    return [int(i) for i in temp]
#     print(p)
```

```python
def check_winner(x):
    if x[0] > x[1]:
        return 'Home Team Wins'
    return 'Away Team Wins'
```

```python
l = penalties.apply(lambda x: convert_to_int(x))
penalty_win = l.apply(lambda x: check_winner(x))
penalty_win
```
